```python
import torch
import matplotlib.pyplot as plt
```

```python
X = torch.arange(-5, 10,0.1, requires_grad=True)
y = torch.relu(X)
X[:5],y[:5]
```

```python
with torch.no_grad():
    plt.plot(X,y)
    plt.grid(True)
```

```python
y.sum().backward(retain_graph=True)
```

```python
with torch.no_grad():
    plt.plot(X, X.grad)
```
