```python
import torch
import matplotlib.pyplot as plt
```

### Relu activation

```python
X = torch.arange(-5, 10,0.1, requires_grad=True)
y = torch.relu(X)
X[:5],y[:5]
```

```python
y.sum().backward(retain_graph=True)
```

```python
with torch.no_grad():
    plt.plot(X,y)
    plt.grid(True)
```

```python
with torch.no_grad():
    plt.plot(X, X.grad)
```

### Sigmoid activation

```python
X.grad.zero_()
```

```python
y = 1/(1 + torch.exp(-X))
```

```python
y
```

```python
y.sum().backward()
```

```python
with torch.no_grad():
    plt.plot(X,y)
```

```python
with torch.no_grad():
    plt.plot(X,X.grad)
```

### Tanh function

```python
X.grad.zero_()
```

```python
y = (1- torch.exp(-2 * X))/ (1 + torch.exp(-2 * X))
```

```python
y.sum().backward()
```

```python
with torch.no_grad():
    plt.plot(X,y, label="x vs y")
    plt.plot(X,X.grad, label="derivative")
    plt.legend()
```

```python
#1

def find_max(X):
    return torch.tensor([x if x>0 else 0 for x in X])

def find_min(X):
    return torch.tensor([x if x<0 else 0 for x in X])

```

```python
alpha = 0.1
y = find_max(X) + alpha * find_min(X)
y
```

```python
with torch.no_grad():
    plt.plot(X,y)
```

```python
y = torch.autograd.Variable(y, requires_grad=True)
```
