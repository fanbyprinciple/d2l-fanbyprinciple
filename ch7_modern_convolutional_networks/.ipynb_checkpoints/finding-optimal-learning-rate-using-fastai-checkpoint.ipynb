### Loading the fastai library

```python
from fastai.vision.all import *
```

### Downloading the dataset.

Cats and dogs in this case.

```python
path = untar_data(URLs.PETS)
```

### Getting the images from path

```python
files = get_image_files(path/"images")

files[:5]
```

### Label function

```python
def label_func(f):
    return f[0].isupper()
```

### Creating a dataloader

```python
dls = ImageDataLoaders.from_name_func(path, files, label_func, item_tfms=Resize(224))
```

### Looking at the dataset

```python
dls.show_batch(max_n=3)
```

Cats will have label True and dogs False.

### Creating a nerual network

```python
learner = cnn_learner(dls, resnet34, metrics=accuracy)
```

### Finding the optimal learning rate automatically

```python
learner.lr_find()
```

### Fine tuning based on optimal learning rate

```python
learner.fine_tune(1, base_lr=9e-3)
```

### Finding best lr again

Since after fine tuning the best lr changes.

```python
learner.lr_find()
```

### Finally training on the lr

```python
learner.fit_one_cycle(1, 7e-7)
```
