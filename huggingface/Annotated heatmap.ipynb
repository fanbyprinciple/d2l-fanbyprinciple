```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

platform = ["Etherium", "AVAX", "Fantom", "Aurora/Near",
              "Polygon", "Optimism", "Arbitrum", "Solana", 
                "Luna", "Atom"]
use = ["Wallets", "Liquidity markets", "Exchanges",
           "Dex Aggregators", "Staking", "Derivative", 
               "Perps/Leverage Trading", "Fund Management", "Options", 
                   "Structured Products", "Insurance", "Yield Farming",
                       "Fixed Interest Rate Protocols", "Bridges", "Gaming",
                           "Stablecoins", "NFT Exchange"]

harvest = np.array([[4, 3, 3, 3, 4, 3, 3, 3, 4, 2],
                    [3, 3, 3, 1, 1, 2, 5, 4, 3, 1],
                    [3, 4, 2, 2, 3, 3, 3, 5, 3, 2],
                    [3, 3, 2, 0, 1 ,2, 2, 1, 0, 0],
                    [4, 3, 1, 1, 1, 0, 1, 3, 2, 1],
                    [1, 2, 0, 1, 0, 2, 0, 1, 2, 3],
                    [3, 2, 0, 0, 1, 2, 1, 2, 1, 0],
                    [4, 0, 0, 0, 1, 3, 0, 3, 1, 0],
                    [1, 1, 0, 0, 1, 2, 3, 4, 3, 0],
                    [3, 2, 1, 1, 1, 0, 2, 3, 0, 0],
                    [1, 0, 0, 0, 1, 0, 0, 0, 1, 0],
                    [2, 3, 2, 0, 1, 0, 3, 4, 3, 0],
                    [3, 0, 0, 0, 0, 0, 1, 1, 0, 0],
                    [3, 4, 0, 3, 0, 4, 4, 2, 3, 0],
                    [0, 2, 1, 0, 4, 0, 0, 4, 2, 1],
                    [5, 3, 0, 0, 1, 0, 2, 1, 1, 0],
                    [5, 4, 2, 1, 1, 1, 1, 3, 4, 1]])

negative_harvest = harvest * -1

print(harvest.shape, negative_harvest.shape)
```

```python
fig, ax = plt.subplots()
#im = ax.imshow(negative_harvest, cmap="plasma")
im = ax.imshow(harvest, cmap="YlOrRd")



fig.set_size_inches(18.5, 10.5)


# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(platform)), labels=platform)
ax.set_yticks(np.arange(len(use)), labels=use)
ax.xaxis.set_tick_params(labeltop=True, pad=40, top=True)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
# for i in range(len(use)):
#     for j in range(len(platform)):
#         text = ax.text(j, i, harvest[i, j],
#                        ha="center", va="center", color="w")

ax.set_title("Activity heatmap of crypto world")
fig.tight_layout()
plt.show()
```
